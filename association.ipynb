```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA = pd.read_csv('/content/Market_Basket_Optimisation.csv', header = None)
transacts = []
for i in range(0, 7501):
  transacts.append([str(DATA.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules = apriori(transacts, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame.nlargest(n= 10, columns='Lift')
```

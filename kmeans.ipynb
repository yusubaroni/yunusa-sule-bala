```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Mall_Customers.csv')
x = data.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  k_means = KMeans(n_clusters=i, init='k-means++', random_state=42)
  k_means.fit(x)
  wcss.append(k_means.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k_means = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_means = k_means.fit_predict(x)
print(y_means)
```

```python
plt.scatter(x[y_means == 0, 0], x[y_means == 0,1], s=100, c='red', label='cluster 1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s=100, c='yellow', label='cluster 1')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s=100, c='green', label='cluster 1')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s=100, c='blue', label='cluster 1')
plt.scatter(x[y_means == 4, 0], x[y_means == 4, 1], s=100, c='magenta', label='cluster 1')
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=300, c='cyan', label='Centroids')
plt.title('cluster')
plt.xlabel('Annual income')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```

```python
prases = Phrases(gbak)
bigram = Phraser(prases)
```

```python
# Convert dictionary items to a list and get the last few
bigram.phrasegrams
```

```python
tokenize_sent = 'jon live in new york city, new york'.split()
tokenize_sent
```

```python
bigram[tokenize_sent]
```

```python
lower_sent = []
for w in gbak:
  lower_sent.append([w.lower() for w in w if w.lower() not in string.punctuation])
```

```python
lower_sent[:4]
```

```python
lower_bigram = Phraser(Phrases(lower_sent))
```

```python
lower_bigram.phrasegrams
```

```python
clean_sent = []
for s in lower_sent:
  clean_sent.append(lower_bigram[s])
```

```python
clean_sent
```

```python
model = Word2Vec(clean_sent, min_count=1, vector_size=100, sg = 1, window=10, epochs = 5, workers = 4)
```

```python
model.save('clean_gutenberg_model.wv')
```
